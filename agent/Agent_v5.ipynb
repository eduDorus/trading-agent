## Calculate new data

```python
def calc_SMA(time_period, database):
    l = len(database)
    SMA = np.zeros(l)
    x = np.zeros(l-time_period+1)
    
    for i in range(time_period):
        x += database[i:l-time_period+1+i]
        
    x /= time_period
    SMA[time_period-1:]=x[:]
    return SMA
```

```python
def calc_EMA(time_period, database):
    l = len(database)
    EMA = np.zeros(l)
    
    SMA = calc_SMA(time_period, database)
    multiplier = (2 / (time_period + 1))
    x = (database[time_period:] - SMA[time_period-1:l-1]) * multiplier + SMA[time_period-1:l-1]
    
    EMA = SMA
    EMA[time_period:] = x[:]
    return EMA
```

```python
def calc_macd(database):
    EMA_12 = calc_EMA(12, database)
    EMA_26 = calc_EMA(26, database)
    
    MACD_line = EMA_12 - EMA_26
    Signal_line = calc_EMA(9, MACD_line)
    MACD_histo = MACD_line - Signal_line
    
    return MACD_line, Signal_line, MACD_histo
```

```python
def calc_avg_gain_loss(time_period, database):
    l = len(database)
    change = np.zeros(l)
    gain = np.zeros(l)
    loss = np.zeros(l)
    avg_gain = np.zeros(l)
    avg_loss = np.zeros(l)
    
    # Calculate change
    change[1:l] = database[1:l]-database[0:l-1]
    gain[change>0] = abs(change[change>0])
    loss[change<0] = abs(change[change<0])
    
    avg_gain[time_period+1] = sum(gain[0:time_period]) / time_period
    avg_loss[time_period+1] = sum(loss[0:time_period]) / time_period
    for i in range(time_period+2,l):
        avg_gain[i] = ((avg_gain[i-1] * (time_period-1)) + gain[i]) / time_period
        avg_loss[i] = ((avg_loss[i-1] * (time_period-1)) + loss[i]) / time_period

    return avg_gain, avg_loss
```

```python
def calc_rsi(time_period, database):
    # Get average gain and loss
    avg_gain, avg_loss =  calc_avg_gain_loss(time_period, database)
    
    l = len(database)
    RSI = np.zeros(l)
    RSI[avg_loss==0] = 100
    index = np.logical_and(avg_loss!=0, avg_gain!=0)
    RSI[index] = 100 - (100 / (1 + (avg_gain[index] / avg_loss[index])))
    return RSI
```

## Define DQN Agent

```python
K.set_image_dim_ordering('tf')

class DQNAgent:
    def __init__(self, frame_size, frame_parameter_number, aux_number, action_size):
        self.frame_size = frame_size
        self.frame_parameter_number =  frame_parameter_number
        self.aux_number = aux_number
        self.action_size = action_size
        self.memory = deque(maxlen=2000)
        self.gamma = 0.95    # discount rate
        self.epsilon = 1.0  # exploration rate
        self.epsilon_min = 0.01
        self.epsilon_decay = 0.995
        self.learning_rate = 0.001
        self.model = self._build_model()

    def _build_model(self):
        # Neural Net for Deep-Q learning Model
        # Main input (the time frame)
        main_input = Input(shape=(self.frame_size,frame_parameter_number))
        main = Conv1D(16, 3, padding='same')(main_input)
        main = BatchNormalization(axis=2)(main)
        main = Activation('relu')(main)
        main = MaxPooling1D(pool_size=2)(main)
        
        main = Conv1D(32, 3, padding='same')(main)
        main = BatchNormalization(axis=2)(main)
        main = Activation('relu')(main)
        main = MaxPooling1D(pool_size=2)(main)
        
        main = Conv1D(64, 3, padding='same')(main)
        main = BatchNormalization(axis=2)(main)
        main = Activation('relu')(main)
        main = MaxPooling1D(pool_size=2)(main)
        
        main = Conv1D(128, 3, padding='same')(main)
        main = BatchNormalization(axis=2)(main)
        main = Activation('relu')(main)
        main = Flatten()(main)
        
        # Additional input (other inputs)
        aux_input = Input(shape=(aux_number,))
        x = keras.layers.concatenate([main, aux_input])
        x = Dense(64, activation='relu')(x)
        x = Dense(64, activation='relu')(x)
        x = Dense(16, activation='relu')(x)
        out = Dense(self.action_size, activation='linear')(x)
        
        model = Model(inputs=[main_input, aux_input], outputs=out)
        model.compile(loss='mse', optimizer=Adam(lr=self.learning_rate))
        print(model.summary())
        return model

    def remember(self, state, action, reward, next_state, done):
        self.memory.append((state, action, reward, next_state, done))

    def act(self, state):
        if np.random.rand() <= self.epsilon:
            # The agent acts randomly
            return random.randrange(self.action_size)
        
        # Predict the reward value based on the given state
        act_values = self.model.predict(state)
        
        # Pick the action based on the predicted reward
        return np.argmax(act_values[0])  # returns action

    def replay(self, batch_size):
        # Sample minibatch from the memory
        minibatch = random.sample(self.memory, batch_size)
        
        # Extract informations from each memory
        for state, action, reward, next_state, done in minibatch:
            
            # if done, make our target reward
            target = reward
            if not done:
                # predict the future discounted reward
                target = (reward + self.gamma *
                          np.amax(self.model.predict(next_state)[0]))
                
            # make the agent to approximately map
            # the current state to future discounted reward
            # We'll call that target_f
            target_f = self.model.predict(state)
            target_f[0][action] = target
            
            # Train the Neural Net with the state and target_f
            self.model.fit(state, target_f, epochs=1, verbose=0)
        if self.epsilon > self.epsilon_min:
            self.epsilon *= self.epsilon_decay

    def load(self, name):
        self.model.load_weights(name)

    def save(self, name):
        self.model.save_weights(name)
```

## Define Trading Game

```python
class TradingGame:
    def __init__(self, start_capital, frame_size, frame_parameter_number, buy_step_size, database):
        self.start_capital = start_capital
        self.frame_size = frame_size
        self.buy_step_size = buy_step_size
        
        self.database = np.asarray(database)
        
        self.capital = start_capital
        self.liquidity = start_capital
        self.timestep = 0
        self.time_frame = self.database[0:self.frame_size]
        
        self.position_price = 0.0
        self.position_amount = 0.0
        self.state = None
    
    ### Reset function
    def reset(self):
        # Set everything to the initial start value
        self.capital = self.start_capital
        self.liquidity = self.start_capital
        self.timestep = 0
        self.time_frame = self.database[0:self.frame_size]
        
        self.position_price = 0.0
        self.position_amount = 0
        
        # state => (time_frame, capital, position)
        frame = np.reshape(self.time_frame, [1, frame_size, frame_parameter_number])
        aux = np.reshape([self.liquidity, self.capital, self.position_amount, self.position_price], [1, 4])
        self.state = [frame, aux]
        
        return self.state
    
    
    ### Next frame function
    def get_next_frame(self):
        self.timestep += 1
        
        if (self.frame_size+self.timestep) < self.database.shape[0]:
            next_frame = self.database[self.timestep:self.frame_size+self.timestep]
            done = False
        else:
            next_frame = self.time_frame
            done = True
            
        return next_frame, done
    
    
    ### Create a long position
    def new_position(self, new_price, amount):
        # Close existing position
        reward = self.close_position(new_price)
        
        # Create new position
        self.position_price = new_price
        self.position_amount = amount
        
        # Update liquidity
        self.liquidity -= self.position_price * abs(self.position_amount)
            
        return reward
    
    
    ### Close position
    def close_position(self, new_price):
        # Close the position
        margin = self.position_amount * (new_price - self.position_price)
        position_value = new_price * abs(self.position_amount)

        # Reset position values
        self.position_price = 0.0
        self.position_amount = 0

        # Update capital and liquidity
        self.liquidity += position_value 
        self.capital += margin
        
        return margin
        
    
    ### Reward function
    def calc_reward(self, action, next_frame):
        # Get new price
        new_price = next_frame[self.frame_size - 1, 2]
        
        # Check the action and calculate the reward
        #hold position
        if action == 0: 
            reward = 0
            #reward = self.position_amount * (new_price - self.position_price)
            
        #new long position
        elif action == 1: 
            if self.position_amount <= 0:
                reward = self.new_position(new_price, 3)
            else:
                reward = -100
                self.capital += reward
                
        #new short position
        elif action == 2: 
            if self.position_amount >= 0:
                reward = self.new_position(new_price, -3)
            else:
                reward = -100
                self.capital += reward
            
        #close position
        elif action == 3: 
            reward = self.close_position(new_price)
            
        return reward
    
    ### Next step function
    def step(self, action):
        # Get next time frame
        next_frame, done = self.get_next_frame()
        
        # Get reward
        reward = self.calc_reward(action, next_frame)
        
        # check if done
        if self.capital <= 0:
            done = True
            
        # Log stats
        #print("L: {}, C: {}, A: {}, P: {}, ACT: {}, R: {}".format(self.liquidity, self.capital, self.position_amount, self.position_price, action, reward))
            
        # Next state
        next_frame = np.reshape(next_frame, [1, frame_size, frame_parameter_number])
        next_aux = np.reshape([self.liquidity, self.capital, self.position_amount, self.position_price], [1, 4])
        next_state = [next_frame, next_aux]
        
        # update state
        self.state = next_state
        
        return next_state, reward, done
    
    def get_score(self):
        return self.capital
    
    def get_position_amount(self):
        return self.position_amount
```

## Train the agent

### Set parameters

```python
# Parameter for Agent
frame_size = 200 # Time frame
frame_parameter_number = 11
aux_number = 4 # Additional parameters
action_size = 4 # hold, long, short, close
buy_step_size = 3

# Parameter for Trading Game
database = train_DB[50:]
batch_size = 128
EPISODES = 2000
CAPITAL = 100000
#POSITION = 0
```

### Training

```python
# Create agent
agent = DQNAgent(frame_size, frame_parameter_number, aux_number, action_size)

# Initialize Environment
env = TradingGame(CAPITAL, frame_size, frame_parameter_number, buy_step_size, database)
```

```python
#Episodes to train
for e in range(EPISODES):
    
    #Create initial state from time frame
    state = env.reset()

    #time frames
    for time_frame in range(0, train_DB.shape[0]):
        
        #Agent takes action
        action = agent.act(state)

        #Calc reward
        next_state, reward, done = env.step(action)
        
        #Remember action
        agent.remember(state, action, reward, next_state, done)
        
        #Override state with next state
        state = next_state
        
        if done:
            print("episode: {}/{}, score: {}, e: {:.2}, tf: {}"
                  .format(e, EPISODES, env.get_score(), agent.epsilon, time_frame))
            break
       
    #Replay
    if len(agent.memory) > batch_size:
        agent.replay(batch_size)

#Save weights
agent.save("./agent-v5.h5")
        
    
```
